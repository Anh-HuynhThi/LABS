```python
from stackapi import StackAPI
from datetime import datetime
SITE = StackAPI('stackoverflow')
```

#### Question 1: Find the questions and answers of last month.

```python
# your code here
fromd = datetime(2022, 9, 10)
tod = datetime(2022, 10, 10)
questions = SITE.fetch('questions', fromdate=fromd, todate=tod)
answers = SITE.fetch('answers', fromdate=fromd, todate=tod)
print(questions)
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
question = SITE.fetch('questions', fromdate=datetime(2022,10,12), todate=datetime(2022,10,13), min=5, sort='votes', tagged='python')['items'][0]
print(question)
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
# your code here
#site = stackAPI5'stackoverflow')
answers = SITE.fetch('answers', ids=[6784, 6473])
print(answers)
```
